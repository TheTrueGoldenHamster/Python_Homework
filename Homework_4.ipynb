```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

 #    Пункт 1
    
def name_search(document_number):
    user_input = input('Введите номер документа: ')
    for element in document_number:
        if user_input == element['number']:
            return print('Владелец документа: ', element['name'])
        else:
            continue
    return print('Документ не найден в базе')
                   
# name_search(documents)

#    Пункт 2

def place_search(document_number):
    user_input = input('Введите номер документа: ')
    for key, value in document_number.items():
        if user_input in value:
            return print('Документ хранится на полке: ', key)
        else:
            continue
    return print('Документ не найден в базе')

# place_search(directories)

#    Пункт 3

def info(means, places):
    for element in documents:
        element.setdefault('place', 0)
        for key, value in directories.items():
            if element['number'] in value:
                element['place'] = key
                print('№: ', element['number'], ', ', 'тип: ', element['type'], ', ', 'владелец: ', element['name'], ', ', 'полка хранения: ', element['place'], sep='')
            else:
                continue
    return        
# info(documents, directories)
        
#    Пункт 4

def new_place(user_command):
    # copy = directories.copy()
    # user_input = input('Введите команду: ')
    while True:
        user_input_2 = input('Введите номер полки: ')
        if user_input_2 not in directories:
            directories.setdefault(user_input_2, [])
            new_places = []
            for key in directories.keys():
                new_places.append(key)
            return print('Полка добавлена. Текущий перечень полок: ', ', '.join(new_places), '.', sep ='')
        else:
            current_places = []
            for key in directories.keys():
                current_places.append(key)
            return print('Такая полка уже существует. Текущий перечень полок: ', ', '.join(current_places), '.', sep ='')
        break

# new_place(directories)

# Пункт 5

def del_place(user_command):
    # copy = directories.copy()
    # user_input = input('Введите команду: ')
    while True:
        user_input_2 = input('Введите номер полки: ')
        if user_input_2 not in directories.keys():
            current_places = []
            for key in directories.keys():
                current_places.append(key)
            return print('Такой полки не существует. Текущий перечень полок: ', ', '.join(current_places), '.', sep ='')
        elif user_input_2 in directories.keys(): 
            for key, value in directories.items():
                if key == user_input_2 and len(value) == 0:
                    del(directories[key])
                    deleted_places = []
                    for key in directories.keys():
                        deleted_places.append(key)
                    return print('Полка удалена. Текущий перечень полок: ', ', '.join(deleted_places), '.', sep ='')
                elif key == user_input_2 and len(value) != 0:
                    current_places_2 = []
                    for key in directories.keys():
                        current_places_2.append(key)
                    return print('На полке есть документs, удалите их перед удалением полки. Текущий перечень полок: ', ', '.join(current_places_2), '.', sep ='')
        break
        
# del_place(directories)

def main():
    while True:
        user_command = input('Введите комманду: ')
        if user_command == 'p':
            print(name_search(documents))
        elif user_command == 's':
            print(place_search(directories))
        elif user_command == 'l':
            print(info(documents, directories))
        elif user_command == 'adc':
            print(new_place(directories))
        elif user_command == 'dc':
            print(del_place(directories))
        elif user_command == 'q':
            print('Программа завершена')
            break     
                  
main()
```
